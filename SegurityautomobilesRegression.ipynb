```python
import numpy as np
import pandas as pd
#import pycaret
from pycaret.regression import *
from ucimlrepo import fetch_ucirepo
# importar LinearRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

automobile = fetch_ucirepo(id=10)
# data (as pandas dataframes)
X = automobile.data.features
y = automobile.data.targets
data = pd.concat([X, y], axis=1)
# Segmentar los datos en TRAIN/TEST
data_train, data_test = train_test_split(data, test_size=0.2, random_state=42)
#utilizar pycaret
data_train.info()
s = setup(data_train, target = 'symboling',session_id =123 )
best = compare_models()
```

```python
plot_model(best,plot = 'feature')
```

```python
plot_model(best,plot = 'error')
```

```python
final_model = finalize_model(best)
predictions = predict_model(final_model,data=data_test)
predictions.head()
test_result = evaluate_model(final_model)
save_model(final_model,'automobiles')
```
