```python
# Importamos las bibliotecas necesarias
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
# Datos de entrada y salida para la compuerta AND
entradas = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
salidas = np.array([0, 0, 0, 1])
# Creamos el modelo de la red neuronal
modelo = Sequential()
modelo.add(Dense(units=1, input_dim=2, activation='sigmoid'))
# Compilamos el modelo
modelo.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Entrenamos la red neuronal
modelo.fit(entradas, salidas, epochs=1000, verbose=0)
# Evaluamos el modelo
puntuacion = modelo.evaluate(entradas, salidas)
print(f"Pérdida: {puntuacion[0]}, Precisión: {puntuacion[1]}")
# Probamos con nuevas entradas
nuevas_entradas = np.array([[1, 1], [0, 0], [1, 1]])
predicciones = modelo.predict(nuevas_entradas)
print("Predicciones para nuevas entradas:")
for i, entrada in enumerate(nuevas_entradas):
    print(f"Entrada {entrada}: {predicciones[i][0]} (aproximadamente)")
# ¡Listo! Ahora tienes una red neuronal que simula la compuerta AND.
```

```python
#COMPUERTA AND
# Importamos las bibliotecas necesarias
from sklearn.neural_network import MLPClassifier
import numpy as np

# Datos de entrada y salida para la compuerta AND
entradas = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
salidas = np.array([0, 0, 0, 1])  # Salida esperada para la compuerta OR

# Creamos el modelo MLPClassifier
modelo = MLPClassifier(hidden_layer_sizes=(3,), activation='relu', solver='adam', max_iter=1000,learning_rate='adaptive')

# Entrenamos el modelo
modelo.fit(entradas, salidas)

# Probamos con nuevas entradas
nuevas_entradas = np.array([[0, 0], [0, 1], [1, 0],[1, 1]])
predicciones = modelo.predict(nuevas_entradas)

print("Predicciones para nuevas entradas:")
for i, entrada in enumerate(nuevas_entradas):
    print(f"Entrada {entrada}: {predicciones[i]}")
```
